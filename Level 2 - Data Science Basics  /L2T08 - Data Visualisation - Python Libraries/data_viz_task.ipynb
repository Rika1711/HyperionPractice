```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load data
car_df = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
sought_manufacturers = car_df[(car_df["Manufacturer"] == "Audi") | (car_df["Manufacturer"] == "Hyundai") | 
(car_df["Manufacturer"] == "Suzuki") | (car_df["Manufacturer"] == "Toyota")]

sns.set_theme(style="darkgrid")
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=sought_manufacturers)
```

```python
# Additionally - for my own understanding: Check if the results are correct.
print(car_df.groupby('Manufacturer')['Rev.per.mile'].mean())
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**  
Suzuki has the highest average of Revs per Mile. But it looks like there is an outlier for the Toyota cars that has even more Revs per Mile.

### 2. Histogram

```python
# Histogram of MPG

sns.histplot(x='MPG.highway', data=car_df, color='skyblue', label='MPG.highway')
sns.histplot(x='MPG.city', data=car_df, color='red', label='MPG.city')
plt.xlabel('Average MPG')
plt.legend()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**  
It is more fuel efficient to drive on the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_df)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
It seems like there is a is an overall trend: The larger the wheelbase, the larger the turning circle gets.

### 4. Barplot

```python
# Barplot for mean horsepower

sns.barplot(x='Type', y='Horsepower', data=car_df, hue='Type')
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**  
A larger car does not have to mean more horsepower.
The bar plot shows that on average the 'large type' cars have the most horsepower. It also shows that 'small type' cars have the least horsepower on average.
However, there are a few points that need to be considered:
- Vans are likely larger than 'large cars', but have less horse power than large, sporty and midsize cars.  
- Sporty cars have a large error bar. That means, that the values are more spread out (within the range of the error bar). 
  There are likely many sporty cars that have high horsepower, higher even than large cars.  
- Midsize cars also have an error bar that indicated that some midsize cars could have more horsepower than large cars.
