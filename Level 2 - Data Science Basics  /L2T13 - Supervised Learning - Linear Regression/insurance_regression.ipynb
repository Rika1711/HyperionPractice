```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# Import the dataframe.
insurance_data = pd.read_csv('insurance.csv')
```

```python
# Inspect the dataframe columns and shape.
print(insurance_data.shape)
insurance_data.head()
```

```python
# Plot a scatter plot.

sns.set_theme()
sns.scatterplot(data=insurance_data, x='age', y='charges')
```

```python
# Fit the model.
X_insurance = insurance_data['age'].values.reshape(-1, 1)
y_insurance = insurance_data['charges'].values

age_model = LinearRegression()
age_model.fit(X_insurance,y_insurance)
y_pred = age_model.predict(X_insurance)
```

```python
# Plot another model, along with the best-fit line.
sns.scatterplot(data=insurance_data, x='age', y='charges')
plt.plot(X_insurance, y_pred)
plt.show()
```
